```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, Bidirectional, LSTM, Dense, Dropout
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from datasets import load_dataset
from sklearn.metrics import classification_report
from sklearn.utils import class_weight
import gradio as gr

# 1. Р—Р°РіСҖСғР·РәР° РҙР°РҪРҪСӢС…
dataset = load_dataset("tweet_eval", "sentiment")
train_texts = dataset['train']['text']
train_labels = np.array(dataset['train']['label'])
test_texts = dataset['test']['text']
test_labels = np.array(dataset['test']['label'])

# 2. Р’СӢСҮРёСҒР»РөРҪРёРө РІРөСҒРҫРІ РәР»Р°СҒСҒРҫРІ (СҮСӮРҫРұСӢ РјРҫРҙРөР»СҢ РҪРө РІСӢРҙР°РІР°Р»Р° СӮРҫР»СҢРәРҫ "РҪРөР№СӮСҖР°Р»СҢРҪРҫ")
weights = class_weight.compute_class_weight('balanced', classes=np.unique(train_labels), y=train_labels)
class_weights_dict = dict(enumerate(weights))

# 3. РўРҫРәРөРҪРёР·Р°СҶРёСҸ (СғРІРөР»РёСҮРёРј СҒР»РҫРІР°СҖСҢ РҙР»СҸ СӮРҫСҮРҪРҫСҒСӮРё)
vocab_size = 15000
max_length = 60
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(train_texts)

train_padded = pad_sequences(tokenizer.texts_to_sequences(train_texts), maxlen=max_length, padding='post')
test_padded = pad_sequences(tokenizer.texts_to_sequences(test_texts), maxlen=max_length, padding='post')

# 4. РЈР»СғСҮСҲРөРҪРҪР°СҸ Р°СҖС…РёСӮРөРәСӮСғСҖР°: BiLSTM + Dropout (РҙР»СҸ РёСҒРәР»СҺСҮРөРҪРёСҸ РҝРөСҖРөРҫРұСғСҮРөРҪРёСҸ)
model = Sequential([
    Embedding(vocab_size, 128, input_length=max_length),
    Bidirectional(LSTM(64, return_sequences=True)),
    Bidirectional(LSTM(32)),
    Dense(64, activation='relu'),
    Dropout(0.5), # Р—Р°СүРёСӮР° РҫСӮ РҝРөСҖРөРҫРұСғСҮРөРҪРёСҸ
    Dense(3, activation='softmax')
])

model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# 5. РһРұСғСҮРөРҪРёРө СҒ СғСҮРөСӮРҫРј РІРөСҒРҫРІ РәР»Р°СҒСҒРҫРІ
print("РқР°СҮРёРҪР°РөРј РҫРұСғСҮРөРҪРёРө СҒ РұР°Р»Р°РҪСҒРёСҖРҫРІРәРҫР№ РІРөСҒРҫРІ...")
model.fit(train_padded, train_labels, epochs=5, validation_data=(test_padded, test_labels),
          class_weight=class_weights_dict, verbose=1)

# 6. РӨСғРҪРәСҶРёСҸ РҙР»СҸ Gradio [cite: 69]
def predict_sentiment(text):
    seq = tokenizer.texts_to_sequences([text])
    padded = pad_sequences(seq, maxlen=max_length, padding='post')
    pred = model.predict(padded)[0]
    classes = ['рҹ”ҙ РқРөРіР°СӮРёРІРҪСӢР№', 'вҡӘ РқРөР№СӮСҖР°Р»СҢРҪСӢР№', 'рҹҹў РҹРҫР·РёСӮРёРІРҪСӢР№']
    return {classes[i]: float(pred[i]) for i in range(3)}

# РҳРҪСӮРөСҖС„РөР№СҒ
demo = gr.Interface(
    fn=predict_sentiment,
    inputs=gr.Textbox(placeholder="РқР°РҝРёСҲРёСӮРө СҮСӮРҫ-СӮРҫ СҸРІРҪРҫ Р·Р»РҫРө РёР»Рё СҖР°РҙРҫСҒСӮРҪРҫРө РҪР° Р°РҪРіР»РёР№СҒРәРҫРј..."),
    outputs=gr.Label(num_top_classes=3),
    title="РҡРҫСҖСҖРөРәСӮРҪСӢР№ Р°РҪР°Р»РёР· СӮРҫРҪР°Р»СҢРҪРҫСҒСӮРё (BiLSTM)",
    description="Р’СӢРҝРҫР»РҪРёР»: Р•РҙРөРҪРҫРІ РҳСҒР»Р°Рј Р–РҫР»РҫРјР°РҪРҫРІРёСҮ. РҳСҒРҝСҖР°РІР»РөРҪР° РҝСҖРҫРұР»РөРјР° РҙРҫРјРёРҪРёСҖРҫРІР°РҪРёСҸ РҪРөР№СӮСҖР°Р»СҢРҪРҫРіРҫ РәР»Р°СҒСҒР°."
)

demo.launch(debug=True)
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
import cv2
import gradio as gr

# 1. РқР°СҒСӮСҖРҫР№РәРё
source_dir = 'pc_dataset'
classes = ['CPU', 'GPU', 'RAM', 'Motherboard']
img_size = (224, 224)

# 2. Р—Р°РіСҖСғР·РәР° РҙР°РҪРҪСӢС…
def load_data(directory):
    x, y = [], []
    for label, cls in enumerate(classes):
        path = os.path.join(directory, cls)
        if not os.path.exists(path): continue
        for f in os.listdir(path):
            try:
                img = cv2.imread(os.path.join(path, f))
                if img is None: continue
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
                x.append(cv2.resize(img, img_size))
                y.append(label)
            except: continue
    return np.array(x), np.array(y)

print("Р—Р°РіСҖСғР·РәР°...")
x_raw, y_raw = load_data(source_dir)
x_proc = tf.keras.applications.mobilenet_v2.preprocess_input(x_raw.astype(np.float32))

# 3. РЎРҫР·РҙР°РөРј РјРҫРҙРөР»СҢ С„СғРҪРәСҶРёРҫРҪР°Р»СҢРҪСӢРј СҒСӮРёР»РөРј (СҚСӮРҫ Р»СғСҮСҲРө РҙР»СҸ Grad-CAM)
base_model = tf.keras.applications.MobileNetV2(input_shape=(224, 224, 3), include_top=False, weights='imagenet')
base_model.trainable = False # Р—Р°РјРҫСҖРҫР·РёРј РҙР»СҸ СҒСӮР°РұРёР»СҢРҪРҫСҒСӮРё РҪР° РјР°Р»РҫРј РәРҫР»РёСҮРөСҒСӮРІРө С„РҫСӮРҫ

inputs = tf.keras.Input(shape=(224, 224, 3))
x = base_model(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
outputs = layers.Dense(len(classes), activation='softmax')(x)
model = tf.keras.Model(inputs, outputs)

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# 4. РһРұСғСҮРөРҪРёРө
print("РһРұСғСҮРөРҪРёРө...")
model.fit(x_proc, y_raw, epochs=10, batch_size=4, verbose=1)

# 5. РӨСғРҪРәСҶРёСҸ РҝСҖРөРҙСҒРәР°Р·Р°РҪРёСҸ
def predict_image(img):
    try:
        if img is None: return None, None

        # РҹРҫРҙРіРҫСӮРҫРІРәР° С„РҫСӮРҫ
        original_img = img.copy()
        if img.shape[-1] == 4: img = cv2.cvtColor(img, cv2.COLOR_RGBA2RGB)
        img_res = cv2.resize(img, img_size)
        img_array = tf.keras.applications.mobilenet_v2.preprocess_input(np.expand_dims(img_res, 0).astype(np.float32))

        # РҹСҖРөРҙСҒРәР°Р·Р°РҪРёРө
        preds = model.predict(img_array)[0]
        res_dict = {classes[i]: float(preds[i]) for i in range(len(classes))}

        # РЈРҝСҖРҫСүРөРҪРҪСӢР№ Grad-CAM (РІ РұР»РҫРәРө try, СҮСӮРҫРұСӢ РҪРө Р»РҫРјР°СӮСҢ РІСҒС‘)
        try:
            last_conv_layer = base_model.get_layer('out_relu')
            grad_model = tf.keras.models.Model([base_model.inputs], [last_conv_layer.output, base_model.output])

            with tf.GradientTape() as tape:
                conv_outputs, _ = grad_model(base_model.inputs) # РұРөСҖРөРј РІСӢС…РҫРҙ СҒ РұР°Р·РҫРІРҫР№ РјРҫРҙРөР»Рё
                # Р’ СҚСӮРҫР№ СғРҝСҖРҫСүРөРҪРҪРҫР№ РІРөСҖСҒРёРё РҝСҖРҫСҒСӮРҫ РҝРҫРәР°Р¶РөРј Р°РәСӮРёРІР°СҶРёСҺ СҒР»РҫРөРІ
                heatmap = tf.reduce_mean(conv_outputs, axis=-1)[0]
                heatmap = tf.maximum(heatmap, 0) / (tf.math.reduce_max(heatmap) + 1e-10)
                heatmap = heatmap.numpy()

            heatmap_img = cv2.resize(heatmap, (original_img.shape[1], original_img.shape[0]))
            heatmap_img = np.uint8(255 * heatmap_img)
            heatmap_color = cv2.applyColorMap(heatmap_img, cv2.COLORMAP_JET)
            heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)
            output_img = cv2.addWeighted(original_img, 0.6, heatmap_color, 0.4, 0)
        except:
            output_img = original_img # Р•СҒР»Рё Grad-CAM РҪРө РІСӢСҲРөР», РҝСҖРҫСҒСӮРҫ РІРөСҖРҪРөРј РҫСҖРёРіРёРҪР°Р»

        return res_dict, output_img
    except Exception as e:
        return {"Error": str(e)}, None

# 6. РҳРҪСӮРөСҖС„РөР№СҒ
demo = gr.Interface(
    fn=predict_image,
    inputs=gr.Image(),
    outputs=[gr.Label(num_top_classes=3), gr.Image()],
    title="РҡР»Р°СҒСҒРёС„РёРәР°СӮРҫСҖ РҹРҡ РәРҫРјРҝР»РөРәСӮСғСҺСүРёС…"
)

demo.launch(debug=True)
```
